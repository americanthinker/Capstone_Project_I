```python
#change column name for ease of reference
df.rename(columns={'What_is_your_current_location__c':'Location'},inplace=True)
```

```python
#drop uneeded columns
df.drop(['state__c','city__c'], axis=1, inplace=True)
```

```python
#drop columns with < 2 pieces of information
df.drop(['Company_that_you_re_interested_in__c', 'Type__c'], axis=1, inplace=True)
df.drop(['Salutation'], axis=1, inplace=True)
```

```python
#remove all "__c" at the end of column names 

def c_remover(col_name):
    if col_name.endswith('__c'):
        col_name = col_name[:-3]
        return col_name
    else:
        return col_name
```

```python
cols = df.columns
new_cols = [c_remover(col) for col in cols]
```

```python
cols == new_cols
```

```python
df.columns = new_cols
```
