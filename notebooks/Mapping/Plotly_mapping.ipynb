```python
import plotly.graph_objects as go

import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/2011_february_us_airport_traffic.csv')
df['text'] = df['airport'] + '' + df['city'] + ', ' + df['state'] + '' + 'Arrivals: ' + df['cnt'].astype(str)
```

```python
vets = pd.read_csv('vets_coords.csv',  encoding="ISO-8859-1")
```

```python
vets['text'] = vets['Location'] + ': ' + vets['Combined'].apply(lambda x: str(x))
```

```python
vets.to_csv('vets_coords.csv', index=False)
```

```python
palette = ['ice         matter      solar       speed       tempo       thermal turbid      armyrose    brbg        earth       fall        geyser      prgn        piyg picnic      portland    puor        rdgy        rdylbu      rdylgn      spectral    tealrose temps       tropic      balance     curl        delta       edge        hsv         icefire       phase       twilight    mrybm       mygbm']
```

```python
palette = palette[0].replace('   ', ' ').replace('  ', ' ').replace('  ', ' ')
```

```python
p = palette.strip().split()
```

```python
for hue in p:
    print(hue)
    fig = go.Figure(data=go.Scattergeo(
            locationmode = 'USA-states',
            lon = vets['longitude'],
            lat = vets['latitude'],
            text = vets['text'],
            mode = 'markers',
            marker = dict(
                size = vets['scaled'],
                opacity = 0.8,
                reversescale = True,
                autocolorscale = False,
                symbol = 'circle',
                line = dict(
                    width=1,
                    color='rgba(102, 102, 102)'
                ),
                colorscale = hue,
                cmin = 0,
                color = vets['Combined'],
                cmax = vets['Combined'].max(),
                colorbar_title="# of Members"
            )))

    fig.update_layout(
            title = 'Elite Meet National Distribution',
            geo = dict(
                scope='usa',
                projection_type='albers usa',
                showland = True,
                landcolor = "rgb(250, 250, 250)",
                subunitcolor = "rgb(217, 217, 217)",
                countrycolor = "rgb(217, 217, 217)",
                countrywidth = 0.5,
                subunitwidth = 0.5
            ),
        )
    fig.show()
```
